```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import matplotlib

%matplotlib inline

ironman_races =  pd.read_csv('data/half_iron_man_races.csv')
ironman_race_results =  pd.read_csv('data/half_iron_man_results.csv')
```

```python
def HMS(seconds, pos=None):
    seconds = int(seconds)
    hours = seconds / 3600
    seconds -= 3600 * hours
    minutes = seconds / 60
    seconds -= 60 * minutes
    
    if hours == 0:
        if minutes == 0:
            return "%ds" % (seconds)
        return "%dm%02ds" % (minutes, seconds)
    return "%dh%02dm" % (hours, minutes)
```

```python
# Sample race
race_name = 'Ironman 70.3 Asia Pacific Championships'

asia_pacific_championships = ironman_race_results[ironman_race_results.name == race_name]

X = asia_pacific_championships.bike_time
Y = asia_pacific_championships.run_time

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1,1,1)

ax.xaxis.set_major_formatter(plt.FuncFormatter(HMS))
ax.yaxis.set_major_formatter(plt.FuncFormatter(HMS))

ax.scatter(X, Y, color="blue", marker="o")

ax.set_title("%s 2013 Corrlation between Bike and Run Splits" % race_name)

ax.set_xlabel("Avg Bike Time = %s" % HMS(np.average(X)))
ax.set_ylabel("Avg Run Time = %s" % HMS(np.average(Y)))
plt.grid(True)


plt.savefig('Bike_Run_split_correlation.png')
```
