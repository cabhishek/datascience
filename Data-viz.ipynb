```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

races = pd.read_csv('data/five_plus_ironman_races.csv').dropna()
ironman_florida = pd.read_csv('data/ironman_florida_m.csv').dropna()

# Get data for a sample athlete
athlete_races = races[(races.athlete_id == 1020) & (races.qualifier_id==7504)].sort('year')

print "Total athlete races =>", len(athlete_races)
```

```python
def HMS(seconds, pos=None):
    seconds = int(seconds)
    hours = seconds / 3600
    seconds -= 3600 * hours
    minutes = seconds / 60
    seconds -= 60 * minutes
    
    if hours == 0:
        if minutes == 0:
            return "%ds" % (seconds)
        return "%dm%02ds" % (minutes, seconds)
    return "%dh%02dm" % (hours, minutes)
```

```python
ironman_florida_outcome = ironman_florida_df[ironman_florida_df.year.apply(lambda x:x in athlete_races.year.tolist())]


fig = plt.figure(figsize=(8, 6))

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(athlete_races.year, athlete_races.final_time,'b', label='Athlete time')

axes.plot(ironman_florida_outcome.year, ironman_florida_outcome.max_final_time, 'r', label='Race Mix time')
axes.plot(ironman_florida_outcome.year, ironman_florida_outcome.min_final_time, 'g', label='Race Min time')
axes.plot(ironman_florida_outcome.year, ironman_florida_outcome.avg_final_time, 'c', label='Race Avg time')

axes.legend(loc='upper right', shadow=True)

x_formatter = plt.ScalarFormatter(useOffset=False)

axes.xaxis.set_major_formatter(x_formatter)
axes.yaxis.set_major_formatter(plt.FuncFormatter(HMS))

# plt.legend([final_time, (final_time, max_final_time)], ["final_time", "max_final_time"])

axes.set_xlabel('race year')
axes.set_ylabel('final time')
axes.set_title('Ironman florida race history');
plt.grid(True)


plt.savefig('Ironman_florida_race_history.png')
```

```python
from funcy import first, last

bins = [18, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70]

ironman_florida_group_data = {}


for year, group in ironman_florida.groupby('year', sort=True):
    age_bins = pd.cut(group.age, bins=bins)
    ironman_florida_group_data[year] = []
    
    for bin_key, bin_group in group.groupby(age_bins):
        ironman_florida_group_data[year].append({'avg_final_time': bin_group.final_time.mean(),
                                                 'avg_cycle_time': bin_group.cycle_time.mean(),
                                                 'avg_swim_time': bin_group.swim_time.mean(),
                                                 'avg_run_time': bin_group.run_time.mean(),
                                                 'age_group': bin_key})
    

year_df = pd.DataFrame(ironman_florida_group_data[2001])


group_split_fig = plt.figure(figsize=(8, 6))

group_race_split_axes = group_split_fig.add_axes([0.1, 0.1, 0.8, 0.8])

group_race_split_axes.plot(year_df.avg_final_time,'k', label='Avg final time')
group_race_split_axes.plot(year_df.avg_cycle_time,'r', label='Avg cycle time')
# group_race_split_axes.plot(year_df.avg_swim_time,'g', label='Avg swim time')
group_race_split_axes.plot(year_df.avg_run_time,'c', label='Avg run time')

group_race_split_axes.yaxis.set_major_formatter(plt.FuncFormatter(HMS))

group_race_split_axes.legend(loc='upper right', shadow=True)

group_race_split_axes.set_xlabel('age group')
group_race_split_axes.set_ylabel('time')
group_race_split_axes.set_title('Ironman florida year 2001');

plt.savefig('ironman_florida_year_2001.png')
```
