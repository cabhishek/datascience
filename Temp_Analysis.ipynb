```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from funcy import first, last

from pprint import pprint
import datetime

%matplotlib inline

bins = [18, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70]

# Read data
louisville_2010 = pd.read_csv('data/Ironman_Louisville_2010.csv').dropna()
louisville_2011 = pd.read_csv('data/Ironman_Louisville_2011.csv').dropna()

def HMS(seconds, pos=None):
    seconds = int(seconds)
    hours = seconds / 3600
    seconds -= 3600 * hours
    minutes = seconds / 60
    seconds -= 60 * minutes
    
    if hours == 0:
        if minutes == 0:
            return "%ds" % (seconds)
        return "%dm%02ds" % (minutes, seconds)
    return "%dh%02dm" % (hours, minutes)

def _top_75_percentile(df, gender):
    return df[(df.m_f == gender) & (df.final_time < df.final_time.quantile(.75))]

def _results_by_age_group(df):
    age_bins= pd.cut(df.age, bins=bins, right=True)      
            
    data = {bin_key:{'avg_final_time': bin_group.final_time.mean(),
                     'avg_bike_time': bin_group.bike_time.mean(),
                     'avg_swim_time': bin_group.swim_time.mean(),
                     'avg_run_time': bin_group.run_time.mean(),
                     'age_group': bin_key} 
            for bin_key, bin_group in df.groupby(age_bins)}
    
    return pd.DataFrame(data.values())

louisville_2010_males = _top_75_percentile(louisville_2010, gender='M')
louisville_2010_females = _top_75_percentile(louisville_2010, gender='F')

louisville_2011_males = _top_75_percentile(louisville_2011, gender='M')
louisville_2011_females = _top_75_percentile(louisville_2011, gender='F')

    
print "louisville_2010_males =>", len(louisville_2010_males)
print "louisville_2010_females =>", len(louisville_2010_females)

print "louisville_2011_males",  len(louisville_2011_males)
print "louisville_2011_females", len(louisville_2011_females)


louisville_2010_males_groups = _results_by_age_group(louisville_2010_males)
louisville_2010_females_groups = _results_by_age_group(louisville_2010_females)

louisville_2011_males_groups = _results_by_age_group(louisville_2011_males)
louisville_2011_females_groups = _results_by_age_group(louisville_2011_females)


fig = plt.figure(figsize=(12, 10))

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])


axes.plot(louisville_2010_females_groups.avg_final_time.dropna().astype(int), 'r', label='2010 Avg Final Time with Temp[Low:73 high:96]')
axes.plot(louisville_2011_females_groups.avg_final_time.dropna().astype(int), 'b', label='2011 Avg Final Time with Temp[Low:70 high:88]')

axes.yaxis.set_major_formatter(plt.FuncFormatter(HMS))

x_formatter = plt.ScalarFormatter(useOffset=True)
axes.xaxis.set_major_formatter(x_formatter)

axes.legend(loc='upper right', shadow=True)

plt.grid(True)

x_axis = 'Age groups 0:(35-40) 1:(60-65) 2:(20, 25) 3:(25,30) 4:(30,35) 5:(50,55) 6:(45,50) 7:(55,60) 8:(65,70) 9:(40,45)'

axes.set_xlabel(x_axis)
axes.set_ylabel('Avg Final Time')
axes.set_title('Ironman louisville [2010 males vs 2011 males]')

axes.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05),
          ncol=3, fancybox=True, shadow=True)

plt.savefig('Ironman_louisville_females_2010_vs_2011.png')
```
