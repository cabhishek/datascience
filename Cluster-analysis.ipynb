```python
import pandas as pd
import numpy as np

from sklearn.cluster import MiniBatchKMeans, KMeans
from sklearn.metrics.pairwise import pairwise_distances_argmin
from sklearn.datasets.samples_generator import make_blobs

import time

import matplotlib.pyplot as plt
import matplotlib
from funcy import *

%matplotlib inline

ironman_races =  pd.read_csv('data/half_iron_man_races.csv')
ironman_race_results =  pd.read_csv('data/half_iron_man_results.csv')

print ironman_races.columns
```

```python
def race_summary():
    summary = []
    
    for race_id, race_group in ironman_race_results.groupby('race_id'):
        top_75_percentile = race_group[
                race_group.final_time < race_group.final_time.quantile(.50)]

        if len(top_75_percentile) == 0: continue
        
        summary.append({'race_id': race_id, 
                        'avg_final_time': top_75_percentile.final_time.mean(), 
                        'athlete_population': len(top_75_percentile)
                      })
    
    return pd.DataFrame(summary)


merged = pd.merge(race_summary(), ironman_races, on='race_id', how='inner')

print merged.columns
print len(merged)
```

```python
from sklearn.cluster import KMeans
import matplotlib.patches as mpatches

columns = ['avg_final_time', 
           'low_temp', 
           'high_temp', 
           'starting_elevation', 
           'gross_elevation_gain', 
           'high_humidity', 
           'low_humidity']

# columns = ['low_temp', 
#            'high_temp']

X = merged[columns].dropna().values


n_clusters = 3

k_means = KMeans(init='k-means++', n_clusters=3, n_init=10)
k_means.fit(X)

k_means_labels = k_means.labels_
k_means_cluster_centers = k_means.cluster_centers_


colors = ['green', 'cyan', 'red']

fig = plt.figure(figsize=(10, 8))

ax = fig.add_subplot(1,1,1)

for k, col in zip(range(n_clusters), colors):

    my_members = k_means_labels == k
        
    cluster_center = k_means_cluster_centers[k]
    
    ax.plot(X[my_members, 0], X[my_members, 1], 'w',
            markerfacecolor=col, marker='*')
    
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,
            markeredgecolor='k', markersize=6)

ax.set_title('Half Ironman race clusters', fontsize=18)

ax.set_xticks(())
ax.set_yticks(())

plt.legend(handles=[mpatches.Patch(label='230 races, 7 data points per race')])

plt.savefig('half_ironman_race_clusters.png')
```
